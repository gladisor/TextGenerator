```python
import gutenbergpy.textget
import nltk
from collections import Counter

text = gutenbergpy.textget.strip_headers(gutenbergpy.textget.get_text_by_id(2600)).decode('utf-8')

## Tolstoy specific punctuation to remove
text = text.replace('-', ' ')
text = text.replace('—', ' ')

paragraphs = nltk.blankline_tokenize(text.lower())

lines_to_remove = paragraphs[0:386]

paragraphs = [par for par in paragraphs if par not in lines_to_remove]
```

```python
tokens = list(map(lambda x: nltk.word_tokenize(x), paragraphs))
```

```python
words = Counter(sum(tokens, []))
```

```python
import matplotlib.pyplot as plt
```

```python
words.most_common()
```

```python
words.add('<right_pad>')
```

```python
print(('<right_pad>' in words))
```

```python
vocabulary = dict(enumerate(words))
```

```python
vocabulary = {v:k for k, v in vocabulary.items()}
```
